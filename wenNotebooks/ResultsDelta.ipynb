```python
import pandas as pd
```

```python
# Export File
path = '../dataFiles/processedData/countypres_2000-2016_processed.csv'
election_data_processed = pd.read_csv(path)
election_data_processed
```

```python
results_2012 = election_data_processed.loc[election_data_processed['year']==2012,:].copy()
results_2016 = election_data_processed.loc[election_data_processed['year']==2016,:].copy()
```

```python
results_2012 = results_2012[['FIPS','state','county','republican_minus_democratic']].reset_index(drop=True)
results_2012.rename(columns={'republican_minus_democratic':'republican_minus_democratic_2012'}, inplace=True)

results_2016 = results_2016[['FIPS','republican_minus_democratic']].reset_index(drop=True)
results_2016.rename(columns={'republican_minus_democratic':'republican_minus_democratic_2016'}, inplace=True)
```

```python
results_2012_to_2016 = pd.merge(results_2012, results_2016[['FIPS','republican_minus_democratic_2016']], how='left', on='FIPS')

results_2012_to_2016.insert(len(results_2012_to_2016.columns), 'change', results_2012_to_2016['republican_minus_democratic_2016']-results_2012_to_2016['republican_minus_democratic_2012'])

# Export File
path = '../dataFiles/processedData/countypres_2012_2016_changes.csv'
results_2012_to_2016.to_csv(path,index=False)
```
