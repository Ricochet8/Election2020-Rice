```python
import pandas as pd
```

```python
# Get File
path = '../dataFiles/rawData/religion_by_county_2010.csv'
religion_data = pd.read_csv(path)
religion_data.head()
```

```python
# Make Sure FIPS are 5 Digits
# Make Int a String
religion_data['FIPS'] = religion_data['FIPS'].astype(str)
# If FIPS String is 4 Digits Long- Add 0 to Beginning

religion_data.insert(len(religion_data.columns), 'Add Zero FIPS', '0' + religion_data['FIPS'])

religion_data.loc[religion_data['FIPS'].str.len()==4, 'FIPS'] = religion_data.loc[religion_data['FIPS'].str.len()==4, 'Add Zero FIPS'] 
```

```python
# Test if There are FIPS That Are Not 5 Characters
test = religion_data.loc[religion_data['FIPS'].str.len()!=5, 'FIPS'].copy()
print(test.size)
```

```python
# Normalize Results To Percentage of Population
religion_data = religion_data[['FIPS', 'Evangelical Protestant', 'Black Protestant', 'Mainline Protestant', 'Catholic',	'Orthodox',	'Other']]
religion_data['Evangelical Protestant'] = religion_data['Evangelical Protestant']/100
religion_data['Black Protestant'] = religion_data['Black Protestant']/100
religion_data['Mainline Protestant'] = religion_data['Mainline Protestant']/100
religion_data['Catholic'] = religion_data['Catholic']/100
religion_data['Orthodox'] = religion_data['Orthodox']/100
religion_data['Other'] = religion_data['Other']/100
```

```python
# Attach Year
religion_data.insert(len(religion_data.columns), 'Year', 2010)
```

```python
# Save Output 
path = '../dataFiles/processedData/religion_by_county_2010_processed.csv'
religion_data.to_csv(path,index=False)
```
