```python
import pandas as pd
```

```python
# Paths
path = '../dataFiles/processedData/countypres_2012_2016_changes.csv'
# Get Data
results_2012_to_2016 = pd.read_csv(path)
```

```python
# Make Election Results FIPS 5 Digits
# Make Sure FIPS are 5 Digits
# Make Int FIPS a String
results_2012_to_2016['FIPS'] = results_2012_to_2016['FIPS'].astype(str)
# If FIPS String is 4 Digits Long- Add 0 to Beginning
results_2012_to_2016.insert(len(results_2012_to_2016.columns), 'Add Zero FIPS', '0' + results_2012_to_2016['FIPS'])
results_2012_to_2016.loc[results_2012_to_2016['FIPS'].str.len()==4, 'FIPS'] = results_2012_to_2016.loc[results_2012_to_2016['FIPS'].str.len()==4, 'Add Zero FIPS']
del results_2012_to_2016['Add Zero FIPS']
results_2012_to_2016
# Make Percents Doubles
results_2012_to_2016['republican_minus_democratic_2012'] = results_2012_to_2016['republican_minus_democratic_2012']*100
results_2012_to_2016['republican_minus_democratic_2016'] = results_2012_to_2016['republican_minus_democratic_2016']*100
results_2012_to_2016['change'] = results_2012_to_2016['change']*100
```

```python
# Get Variables from 2016
path = '../dataFiles/mergedData/election_results_with_variables_no_blanks.xlsx'
election_results_with_variables_no_blanks = pd.read_excel(path, sheet_name='Sheet1')
election_results_with_variables_no_blanks_2016 = election_results_with_variables_no_blanks.loc[election_results_with_variables_no_blanks['year'] == 2016].copy()
# Make Sure FIPS are 5 Digits
# Make Int FIPS a String
election_results_with_variables_no_blanks_2016['FIPS'] = election_results_with_variables_no_blanks_2016['FIPS'].astype(str)
# If FIPS String is 4 Digits Long- Add 0 to Beginning
election_results_with_variables_no_blanks_2016.insert(len(election_results_with_variables_no_blanks_2016.columns), 'Add Zero FIPS', '0' + election_results_with_variables_no_blanks_2016['FIPS'])
election_results_with_variables_no_blanks_2016.loc[election_results_with_variables_no_blanks_2016['FIPS'].str.len()==4, 'FIPS'] = election_results_with_variables_no_blanks_2016.loc[election_results_with_variables_no_blanks_2016['FIPS'].str.len()==4, 'Add Zero FIPS']
del election_results_with_variables_no_blanks_2016['Add Zero FIPS']
election_results_with_variables_no_blanks_2016.head()
```

```python
columns_to_keep_original =  [
'FIPS',
'Evangelical Protestant',
'Black Protestant',
'Mainline Protestant',
'Catholic',
'Orthodox',
'Other Religion',
'Non Religious %',     
'Population Density (Per Sq. Mile)',
'% Total Population: Male',
'% Total Population: Female',
'% Total Population: 18 to 34 Years',
'% Total Population: 35 to 64 Years',
'% Total Population: 65 and Over',
'% Total Population: White Alone',
'% Total Population: Black or African American Alone',
'% Total Population: American Indian and Alaska Native Alone',
'% Total Population: Asian Alone',
'% Total population: Hispanic or Latino',
'% Population 15 Years and Over: Never Married',
'% Population 15 Years and Over: Now Married (Not Including Separated)',
'% Population 15 Years and Over: Divorced',
'% Single Parent Households',
'Average Household Size',
'College or Above',
'Gini Index',
'Median Income',
'% Civilian Population 18 Years and Over: Veteran',
'% Civilian Population 18 Years and Over: Nonveteran',
]
```

```python
election_results_with_variables_no_blanks_2016 = election_results_with_variables_no_blanks_2016[columns_to_keep_original]
results_2012_to_2016 = pd.merge(results_2012_to_2016, election_results_with_variables_no_blanks_2016, on='FIPS', how='left')
```

```python
# Export Final Dataframe
output_path  = '../dataFiles/mergedData/election_results_2012_to_2016_changes_with_variables.xlsx'
results_2012_to_2016.to_excel(output_path,index=False)

# Save Results With NA Dropped
results_2012_to_2016_no_blanks = results_2012_to_2016.dropna().copy().reset_index(drop=True)
output_path  = '../dataFiles/mergedData/election_results_2012_to_2016_changes_with_variables_no_blanks.xlsx'
results_2012_to_2016_no_blanks.to_excel(output_path,index=False)
```
