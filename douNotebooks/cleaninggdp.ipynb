```python
import pandas as pd
```

```python
# Get File
path = r'C:\Users\Owner\Documents\School\Homework\Election2020-Rice\douNotebooks\GDP_per_capita_2012_2015.csv'
gdppc_df = pd.read_csv(path)
gdppc_df.head()
```

```python
# gdppc_df0 = gdppc_df.rename(columns=gdppc_df.iloc[0])
# gdppc_df0.drop(gdppc_df.index[0]).head()
gdppc_df.dtypes
gdppc_df.dropna(inplace=True)
```

```python
# Make Sure FIPS are 5 Digits
# Make Int a String, already a string???
gdppc_df['Info'] = gdppc_df['Info'].astype(str)
#Drop random .0's 
#df['ID'] = df['ID'].astype(str).replace('\.0', '', regex=True)
gdppc_df['Info'] = gdppc_df['Info'].replace('\.0', '', regex=True)
```

```python
# If FIPS String is 4 Digits Long- Add 0 to Beginning

gdppc_df.insert(len(gdppc_df.columns), 'Add Zero FIPS', '0' + gdppc_df['Info'])

gdppc_df.loc[gdppc_df['Info'].str.len()==4, 'Info'] = gdppc_df.loc[gdppc_df['Info'].str.len()==4, 'Add Zero FIPS'] 
```

```python
# Test if There are FIPS That Are Not 5 Characters
test = gdppc_df.loc[gdppc_df['Info'].str.len()!=5, 'Info'].copy()
print(test.size)
```
