```python
import pandas as pd
import numpy as np
```

```python
data2018_to_load = "../Education/ACSDT1Y2018.B07009_data_with_overlays_2019-12-21T130141.csv"
metadata2018_to_load = "../Education/ACSDT1Y2018.B07009_metadata_2019-12-21T130141.csv"

fulldata2018 = pd.read_csv(data2018_to_load)
metadata2018 = pd.read_csv(metadata2018_to_load)
```

```python
fulldata2018.head()
```

```python
data2018 = fulldata2018[['GEO_ID',"NAME","B07009_003E","B07009_004E","B07009_005E","B07009_006E","B07009_001E"]]

data2018 = data2018.rename(columns={"GEO_ID":"FIPS","NAME":"County",
                                        "B07009_002E":"Less Than High School Graduate",
                                        "B07009_003E":"High school graduate (includes equivalency)",
                                        "B07009_004E":"Some college or associate's degree",
                                        "B07009_005E":"Bachelor's degree",
                                        "B07009_006E":"Graduate or professional degree","B07009_001E":"Total"})
```

```python
data2018.head()
```

```python
data2018=data2018.drop(index =0)
```

```python
data2018=data2018.reset_index(drop=True)
```

```python
data2018.head()
```

```python
countySplit = data2018["County"].str.split(", ", n = 1, expand = True) 
```

```python
countySplit.head()
```

```python
data2018["County"]= countySplit[0] 
data2018["State"]=countySplit[1]
```

```python
data2018
```

```python
countyDelete = data2018["County"].str.split(" ", n = 1, expand = True) 
```

```python
data2018["County"]= countyDelete[0] 
data2018["Delete"]=countyDelete[1]
```

```python
data2018
```

```python
data2018.drop(columns=['Delete'])
```

```python
splitFIPS = data2018["FIPS"].str.split("US", n = 1, expand = True) 
```

```python
splitFIPS.head()
```

```python
data2018["Delete"]= splitFIPS[0] 
data2018["FIPS"]=splitFIPS[1]
```

```python
data2018.drop(columns=['Delete'], inplace = True)
```

```python
data2018.insert(len(data2018.columns), 'Year', 2018)
```

```python
data2018.head()
```

```python
path = '../dataPoints/education_2018_clean.csv'
data2018.to_csv(path,index=False)
```
