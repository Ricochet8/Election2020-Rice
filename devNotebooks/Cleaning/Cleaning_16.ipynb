```python
import pandas as pd
import numpy as np
```

```python
data2016_to_load = "../Education/ACSDT1Y2016.B07009_data_with_overlays_2019-12-21T130141.csv"
metadata2016_to_load = "../Education/ACSDT1Y2016.B07009_metadata_2019-12-21T130141.csv"

fulldata2016 = pd.read_csv(data2016_to_load)
metadata2016 = pd.read_csv(metadata2016_to_load)
```

```python
fulldata2016.head()
```

```python
data2016 = fulldata2016[['GEO_ID',"NAME","B07009_003E","B07009_004E","B07009_005E","B07009_006E","B07009_001E"]]

data2016 = data2016.rename(columns={"GEO_ID":"FIPS","NAME":"County",
                                        "B07009_002E":"Less Than High School Graduate",
                                        "B07009_003E":"High school graduate (includes equivalency)",
                                        "B07009_004E":"Some college or associate's degree",
                                        "B07009_005E":"Bachelor's degree",
                                        "B07009_006E":"Graduate or professional degree","B07009_001E":"Total"})
```

```python
data2016.head()
```

```python
data2016=data2016.drop(index =0)
```

```python
data2016=data2016.reset_index(drop=True)
```

```python
data2016.head()
```

```python
countySplit = data2016["County"].str.split(", ", n = 1, expand = True) 
```

```python
countySplit.head()
```

```python
data2016["County"]= countySplit[0] 
data2016["State"]=countySplit[1]
```

```python
data2016
```

```python
countyDelete = data2016["County"].str.split(" ", n = 1, expand = True) 
```

```python
data2016["County"]= countyDelete[0] 
data2016["Delete"]=countyDelete[1]
```

```python
data2016
```

```python
data2016.drop(columns=['Delete'])
```

```python
splitFIPS = data2016["FIPS"].str.split("US", n = 1, expand = True) 
```

```python
splitFIPS.head()
```

```python
data2016["Delete"]= splitFIPS[0] 
data2016["FIPS"]=splitFIPS[1]
```

```python
data2016.drop(columns=['Delete'], inplace = True)
```

```python
data2016.insert(len(data2016.columns), 'Year', 2016)
```
