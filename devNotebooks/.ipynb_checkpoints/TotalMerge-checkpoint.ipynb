```python
import pandas as pd
import numpy as np
```

```python
data2010_to_load = "dataPoints/education_2010_clean.xlsx"
data2011_to_load = "dataPoints/education_2011_clean.xlsx"
data2012_to_load = "dataPoints/education_2012_merged.xlsx"
data2013_to_load = "dataPoints/education_2013_clean.xlsx"
data2014_to_load = "dataPoints/education_2014_clean.xlsx"
data2015_to_load = "dataPoints/education_2015_clean.xlsx"
data2016_to_load = "dataPoints/education_2016_merged.xlsx"
data2017_to_load = "dataPoints/education_2017_clean.xlsx"
data2018_to_load = "dataPoints/education_2018_merged.xlsx"

data2010 = pd.read_excel(data2010_to_load)
data2011 = pd.read_excel(data2011_to_load)
data2012 = pd.read_excel(data2012_to_load)
data2013 = pd.read_excel(data2013_to_load)
data2014 = pd.read_excel(data2014_to_load)
data2015 = pd.read_excel(data2015_to_load)
data2016 = pd.read_excel(data2016_to_load)
data2017 = pd.read_excel(data2017_to_load)
data2018 = pd.read_excel(data2018_to_load)
```

```python
years = [data2010, data2011, data2012, data2013, data2014, data2015, data2016, data2017, data2018]
```

```python
allData = pd.concat(years)
```

```python
allData
```

```python
#allData = allData.dropna(inplace = True)
allData.dropna(inplace = True)
```

```python
allData = allData[allData.County != 'United']
```

```python
allData.dtypes
```

```python
allData['FIPS'] = allData['FIPS'].astype(str)
```

```python
allData['FIPS'] = allData['FIPS'].replace('\.0','', regex = True)
allData.head()
```

```python
allData.insert(len(allData.columns), 'Add Zero FIPS', '0' + allData['FIPS'])
allData.loc[allData['FIPS'].str.len()==4, 'FIPS'] = allData.loc[allData['FIPS'].str.len()==4, 'Add Zero FIPS']
del allData['Add Zero FIPS']
```

```python
allData
```

```python
path = '../dataFiles/processedData/education_processed.xlsx'
allData.to_excel(path,index=False)
```
