```python
import pandas as pd
```

```python
# Get File
path = 'Desktop/Election/5yearmedianincome_county.csv'
income_df = pd.read_csv(path)
income_df.head()
```

```python
income_df.dtypes
income_df.dropna(inplace=True)
```

```python
# Make Sure FIPS are 5 Digits
# Make Int a String, already a string???
income_df['FIPS'] = income_df['FIPS'].astype(str)
#Drop random .0's 
#df['ID'] = df['ID'].astype(str).replace('\.0', '', regex=True)
income_df['FIPS'] = income_df['FIPS'].replace('\.0', '', regex=True)
income_df.head()
```

```python
# If FIPS String is 4 Digits Long- Add 0 to Beginning

income_df.insert(len(income_df.columns), 'Add Zero FIPS', '0' + income_df['FIPS'])

income_df.loc[income_df['FIPS'].str.len()==4, 'FIPS'] = income_df.loc[income_df['FIPS'].str.len()==4, 'Add Zero FIPS'] 
```

```python
# Test if There are FIPS That Are Not 5 Characters
test = income_df.loc[income_df['FIPS'].str.len()!=5, 'FIPS'].copy()
print(test.size)
```

```python
#Clean columns
#df = df.rename(columns={'oldName1': 'newName1', 'oldName2': 'newName2'})
income_df = income_df.rename(columns={'Total households!!Median household income (dollars)': 'Household Median Income', 'Margin of Error!!INCOME AND BENEFITS (IN 2018 INFLATION-ADJUSTED DOLLARS)!!Total households!!Median household income (dollars)': 'Margin of Error'})
income_df.head()
```

```python
income_df = income_df[['FIPS', 'Household Median Income', 'Margin of Error']]
#income_df.dropna(inplace=True)
#income_df
```

```python
# Attach Year
#income_df.insert(len(income_df.columns), 'Year', 2018)
```

```python
# Save Output 
path2 = 'Desktop/Election/medianincome_processed.csv'
income_df.to_csv(path2,index=False)
```

```python
path3 = 'Desktop/Election/2012_income.csv'
income_2012 = pd.read_csv(path3)
path4 = 'Desktop/Election/2016_income.csv'
income_2016 = pd.read_csv(path4)
path5 = 'Desktop/Election/2018_income.csv'
income_2018 = pd.read_csv(path5)
income_2012.dropna(inplace=True)
income_2016.dropna(inplace=True)
income_2018.dropna(inplace=True)
income_2012['FIPS'] = income_2012['FIPS'].astype(str)
income_2016['FIPS'] = income_2016['FIPS'].astype(str)
income_2018['FIPS'] = income_2018['FIPS'].astype(str)
```

```python
income_2012['FIPS'] = income_2012['FIPS'].replace('\.0', '', regex=True)
income_2016['FIPS'] = income_2016['FIPS'].replace('\.0', '', regex=True)
income_2018['FIPS'] = income_2018['FIPS'].replace('\.0', '', regex=True)
income_2012.insert(len(income_2012.columns), 'Add Zero FIPS', '0' + income_2012['FIPS'])
income_2012.loc[income_2012['FIPS'].str.len()==4, 'FIPS'] = income_2012.loc[income_2012['FIPS'].str.len()==4, 'Add Zero FIPS']
income_2016.insert(len(income_2016.columns), 'Add Zero FIPS', '0' + income_2016['FIPS'])
income_2016.loc[income_2016['FIPS'].str.len()==4, 'FIPS'] = income_2016.loc[income_2016['FIPS'].str.len()==4, 'Add Zero FIPS']
income_2018.insert(len(income_2018.columns), 'Add Zero FIPS', '0' + income_2018['FIPS'])
income_2018.loc[income_2018['FIPS'].str.len()==4, 'FIPS'] = income_2018.loc[income_2018['FIPS'].str.len()==4, 'Add Zero FIPS']
```

```python
income_2012 = income_2012.rename(columns={'Mean household income (dollars)': 'Median Household income 2012'})
income_2012 = income_2012[['FIPS', 'Median Household income 2012']]
income_2016 = income_2016.rename(columns={'Total households!!Median household income (dollars)': 'Median Household income 2016'})
income_2016 = income_2016[['FIPS', 'Median Household income 2016']]
income_2018 = income_2018.rename(columns={'Total households!!Median household income (dollars)': 'Median Household income 2018'})
income_2018 = income_2018[['FIPS', 'Median Household income 2018']]
```

```python
#final_df = pd.merge(NYC_df, JC_df, on = columns, how= 'outer')
merge1 = pd.merge(income_2012, income_2016, on=['FIPS'], how='outer')
merge2 = pd.merge(merge1, income_2018, on=['FIPS'], how='outer')
merge2.dropna()
election_year_income = merge2
election_year_income.head()
```

```python
path8 = 'Desktop/Election/election_year_income_processed.csv'
election_year_income.to_csv(path8,index=False)
```
